# Task Implementation

The given first dynamical system is described by the differential equation:

$\dot{x} = \alpha - x^2$

For a steady state $\dot{x} = 0$: Setting $\alpha - x^2 = 0$, gives ready states $x_0 = \pm \sqrt{\alpha}$, if we solve the equation.

- For $\alpha > 0$: Has two steady states at $x_0 = \pm \sqrt{\alpha}$
- And for $\alpha < 0$: there are no steady states

The given second dynamical system is described by the differential equation:

$\dot{x} = \alpha - 2x^2 -3$

For a steady state $\dot{x} = 0$: Setting $\alpha - 2x^2 -3$, gives ready states $x_0 = \pm \sqrt{\frac{\alpha - 3}{2}}$, if we solve the equation.

- For $\alpha-3 > 0$: Has two steady states at $x_0 = \pm \sqrt{\frac{\alpha - 3}{2}}$
- And for $\alpha < 0$: there are no steady states



```python
def find_steady_states_first_system(alpha):
    if alpha > 0:
        return np.array([np.sqrt(alpha), -np.sqrt(alpha)])
    else:
        return np.array([])


def find_steady_states_second_system(alpha):
    if alpha - 3 > 0:
        return np.array([np.sqrt((alpha - 3) / 2), -np.sqrt((alpha - 3) / 2)])
    else:
        return np.array([])


def plot_bifurcation_diagram(alpha_interval, plot_title, bifurcation_point, find_steady_states_method):
    alpha_values = np.linspace(
        alpha_interval[0], alpha_interval[1], alpha_interval[2])

    x_values_positives = np.zeros_like(alpha_values)
    x_values_negatives = np.zeros_like(alpha_values)

    for i, alpha in enumerate(alpha_values):
        steady_states = find_steady_states_method(alpha)
        if len(steady_states) > 0:
            x_values_positives[i] = steady_states[0]
            x_values_negatives[i] = steady_states[1]

    fig, axs = plt.subplots(1, 2, figsize=(12, 5))

    axs[0].plot(alpha_values, x_values_positives, label='Stable', color='blue')
    axs[0].axvline(x=bifurcation_point, linestyle='--', color='red',
                   label='Bifurcation Point (α={})'.format(bifurcation_point))

    axs[0].set_xlim([alpha_interval[0], alpha_interval[1]])
    axs[0].set_ylim([-1, 1])
    axs[0].set_xlabel('α')
    axs[0].set_ylabel('Steady State (x)')
    axs[0].set_title(r'Bifurcation Dynamic Positive System for ' + plot_title)
    axs[0].legend(loc='upper left')

    axs[1].plot(alpha_values, x_values_negatives,
                label='Unstable', color='green')
    axs[1].axvline(x=bifurcation_point, linestyle='--', color='red',
                   label='Bifurcation Point (α={})'.format(bifurcation_point))

    axs[1].set_xlim([alpha_interval[0], alpha_interval[1]])
    axs[1].set_ylim([-1, 1])
    axs[1].set_xlabel('α')
    axs[1].set_ylabel('Steady State (x)')
    axs[1].set_title(r'Bifurcation Dynamic Negative System for ' + plot_title)
    axs[1].legend(loc='upper left')
```

## Visualization

```python
plot_bifurcation_diagram(
    [-1, 1, 100], r'$\bar{x}= \alpha - x^2$', 0, find_steady_states_first_system)
```

```python
plot_bifurcation_diagram(
    [-3, 5, 100], r'$\bar{x}= \alpha - 2x^2 - 3$', 3, find_steady_states_second_system)
```
